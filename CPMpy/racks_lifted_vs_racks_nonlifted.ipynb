```python
from racks_lifted import LiftedRacksExample
from racks_model import RackConfiguration

(A,B,C,D)=(40,0,0,0)
lr: LiftedRacksExample = LiftedRacksExample(24,100)
lr.model += lr.OBJECTCOUNT[RackConfiguration.ELEMENTA] == A
lr.model += lr.OBJECTCOUNT[RackConfiguration.ELEMENTB] == B
lr.model += lr.OBJECTCOUNT[RackConfiguration.ELEMENTC] == C
lr.model += lr.OBJECTCOUNT[RackConfiguration.ELEMENTD] == D
estimated_frames = 4 + (A + B*3 + C*3 + D*4) // 5
lr.model += (lr.OBJECTCOUNT[RackConfiguration.FRAME] <= estimated_frames)
lr.timelimit = 600
lr.solve_and_show_solution()
```

```python
from racks import RacksExample
from racks_model import RackConfiguration

(A,B,C,D)=(40,0,0,0)
nrofelements = A + B + C + D
nrofmodules = A + 3*B + 3*C + 5*D
nrofframes = 4 + nrofmodules // 5
nrofracks = 1 + nrofframes // 4
size = nrofelements + nrofmodules + nrofframes + nrofracks

r: RacksExample = RacksExample(size)
r.model += r.OBJECTCOUNT[RackConfiguration.ELEMENTA] == A
r.model += r.OBJECTCOUNT[RackConfiguration.ELEMENTB] == B
r.model += r.OBJECTCOUNT[RackConfiguration.ELEMENTC] == C
r.model += r.OBJECTCOUNT[RackConfiguration.ELEMENTD] == D
estimated_frames = 4 + (A + B*3 + C*3 + D*4) // 5
r.model += (r.OBJECTCOUNT[RackConfiguration.FRAME] <= estimated_frames)
r.timelimit = 300
r.solve_and_show_solution()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

def show_performance_results_for_lines(lines):
    nrofobjects_lifted_ok = [ int(line[0]) for line in lines if eval(line[5]) is True ]
    lifted_time_ok = [ float(line[6]) for line in lines if eval(line[5]) is True ]
    nrofobjects_lifted_nok = [ int(line[0]) for line in lines if eval(line[5]) is False ]
    lifted_time_nok = [ float(line[6]) for line in lines if eval(line[5]) is False ]
    plt.scatter(nrofobjects_lifted_ok,lifted_time_ok,c="blue",marker='^',label="lifted")
    plt.scatter(nrofobjects_lifted_nok,lifted_time_nok,c="blue",marker='x',label="lifted (timeout)")
    nrofobjects_nonlifted_ok = [ int(line[0]) for line in lines if eval(line[7]) is True ]
    nonlifted_time_ok = [ float(line[8]) for line in lines if eval(line[7]) is True ]
    nrofobjects_nonlifted_nok = [ int(line[0]) for line in lines if eval(line[7]) is False ]
    nonlifted_time_nok = [ float(line[8]) for line in lines if eval(line[7]) is False ]
    plt.scatter(nrofobjects_nonlifted_ok,nonlifted_time_ok,c="orange",marker='o',label="nonlifted")
    plt.scatter(nrofobjects_nonlifted_nok,nonlifted_time_nok,c="orange",marker='_',label="nonlifted (timeout)")
    plt.title('lifted vs nonlifted')
    plt.xlabel("nr of elements")
    plt.ylabel('Time')
    plt.legend()
    plt.show()

def show_performance_results(fname):
    with open(fname, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        lines = list(csv_reader)
        show_performance_results_for_lines(lines)


show_performance_results('performance_elements_2023041918_60_23.csv')
```
